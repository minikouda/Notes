```python
import copy
import numpy as np
import matplotlib.pyplot as plt
```

In this problem, we solve the following ridge regression problem:
$$\min_{x \in \mathbb{R}^{n}}f_{\lambda}(x) \qquad \text{where} \qquad f_{\lambda}(x) \doteq \frac{1}{2}\left\{\frac{1}{m}\|Ax - y\|_{2}^{2} + \lambda \|x\|_{2}^{2}\right\}$$
via gradient descent with step size $\eta > 0$.

# Computation Functions

```python
def make_f(A, y, lmbda):
    m, n = A.shape
    return lambda x: 0.5 * ((np.linalg.norm(A @ x - y) ** 2) / m + lmbda * np.linalg.norm(x) ** 2)
```

```python
def ridge_regression_sol(A, y, lmbda):
    m, n = A.shape
    # Closed-form ridge regression solution to compare to
    return np.linalg.inv(A.T @ A + lmbda * m * np.eye(A.shape[1])) @ A.T @ y
```

```python
def sigma_max(A):
    S = np.linalg.svd(A, full_matrices=True, compute_uv=False)
    return S[0]

def sigma_min(A):
    S = np.linalg.svd(A, full_matrices=True, compute_uv=False)
    if S.shape[0] < A.shape[1]:
        return np.float64(0.0)
    else:
        return S[-1]
```

```python
def gd(x0, grad_fn, eta, T):
    x_history = []
    x = x0
    for t in range(T):
        x_history.append(copy.deepcopy(x))
        x -= grad_fn(x) * eta
    return np.array(x_history)
```

```python
def eta_opt(A, lmbda):
    m, n = A.shape
    return 2 * m / (sigma_max(A) ** 2 + sigma_min(A) ** 2 + 2 * lmbda * m)
```

```python
def ridge_regression_gd(A, y, lmbda, eta, T):
    m, n = A.shape
    def grad_fn(x):
        return  A.T @ (A @ x - y) / m + lmbda * x
        return ...  # TODO: gradient of our ridge objective
    x0 = np.random.standard_normal((n,))
    return gd(x0, grad_fn, eta, T)
```

```python
def ridge_regression_sgd(A, y, lmbda, eta, T):
    m, n = A.shape

    def grad_fn(x):
        idx = np.random.randint(0, m)
        a_i = A[idx]
        # Unbiased estimate of grad[(1/2m)||Ax-y||^2 + (lambda/2)||x||^2]
        return a_i * (a_i @ x - y[idx]) + lmbda * x

    x0 = np.random.standard_normal((n,))
    return gd(x0, grad_fn, eta, T)
```

```python
def time_average(x_history):
    time_sum = np.cumsum(x_history, axis=0)
    time_avg = time_sum / (1 + np.arange(x_history.shape[0]).reshape(-1, 1))
    return time_avg
```

# Plotting Functions

```python
def plot_data_1d(A, y):
    plt.xlabel("$a$")
    plt.ylabel("$b$")
    plt.plot(A[:, 0], y, marker="o", linestyle="none")
```

```python
def plot_data_2d(A, y):
    ax = plt.figure().add_subplot(projection='3d')
    ax.set_xlabel("$a_{1}$")
    ax.set_ylabel("$a_{2}$")
    ax.plot(A[:, 0], A[:, 1], zs=y, marker="o", linestyle="none")
```

```python
def plot_loss_curves_1d(A, y, lmbdas, x_star):
    m, n = A.shape
    assert n == 1
    for lmbda in lmbdas:
        f = make_f(A, y, lmbda)
        x_range = np.linspace(start=x_star - 10, stop=x_star + 10, num=1000).reshape(-1, 1)
        fx_range = np.array([f(x) for x in x_range])

        x_RR = ridge_regression_sol(A, y, lmbda)
        fx_RR = f(x_RR)

        plt.xlabel("$x$")
        plt.ylabel(r"$f_{\lambda}(x)$")
        plt.plot(x_range, fx_range, label = r"$\lambda = %s$"%lmbda)

        plt.plot(x_RR, fx_RR, 'o')
    plt.legend()
```

```python
def plot_history_lines_1d(A, y, x_history, x_star, ts, time_avg=False):
    plt.xlabel("$a$")
    plt.ylabel("$b$")
    plt.plot(A[:, 0], y, ".")

    if time_avg:
        x_history = time_average(x_history)

    for t in ts:
        x_t = x_history[t-1][0]
        plt.plot([-1, 1], [-x_t, x_t], label="$x_{%s}$"%t)
    plt.plot([-1, 1], [-x_star, x_star], label=r"$x_{\lambda}^{\star}$")
    plt.legend()
```

```python
def plot_x_trajectory_1d(f, x_history, x_star, time_avg=False):
    if time_avg:
        x_history = time_average(x_history)

    x_range = np.linspace(start=x_star - 3, stop=x_star + 3, num=1000).reshape(-1, 1)
    fx_range = np.array([f(x) for x in x_range])

    plt.xlabel("$x$")
    plt.ylabel(r"$f_{\lambda}(x)$")
    plt.plot(x_range, fx_range)

    plt.plot(x_star, f(x_star), 'o')

    f_history = np.array([f(x) for x in x_history]).reshape(-1, 1)

    plt.quiver(x_history[:-1], f_history[:-1],\
           x_history[1:]-x_history[:-1], f_history[1:]-f_history[:-1],\
           scale_units='xy', angles='xy', scale=1)
```

```python
def plot_x_trajectory_2d(f, x_history, x_star, time_avg=False):
    if time_avg:
        x_history = time_average(x_history)

    plt.xlabel("$x_{1}$")
    plt.ylabel("$x_{2}$")

    plt.plot(x_star[0], x_star[1], 'o')

    plt.quiver(x_history[:-1,0], x_history[:-1,1],\
           x_history[1:,0]-x_history[:-1,0], x_history[1:,1]-x_history[:-1,1],\
           scale_units='xy', angles='xy', scale=1)

    xmin, xmax, ymin, ymax = plt.axis()
    x_range = np.linspace(start=xmin, stop=xmax, num=100)
    y_range = np.linspace(start=ymin, stop=ymax, num=100)
    xv, yv = np.meshgrid(x_range, y_range)
    z = np.stack((xv, yv), axis=-1)
    fz = np.zeros((z.shape[0], z.shape[1]))
    for i in range(z.shape[0]):
        for j in range(z.shape[1]):
            fz[i][j] = f(z[i][j])
    min_fz = np.min(fz)
    max_fz = np.max(fz)
    level1 = min_fz + 0.05 * (max_fz - min_fz)
    level2 = min_fz + 0.15 * (max_fz - min_fz)
    plt.contour(x_range, y_range, fz, levels=[level1, level2])
```

```python
def plot_convergence_x(A, x_history, x_star, time_avg=False):
    if time_avg:
        x_history = time_average(x_history)

    x_history = x_history.T  # now it's a (N x T) matrix where each column is an x_t

    plt.xlabel("$t$")
    plt.ylabel(r"$|x_{t} - x_{\lambda}^{\star}|_{i}$")
    plt.yscale("log")
    for i in range(A.shape[1]):
        dist_xi = np.abs(x_history[i] - x_star[i])
        plt.plot(dist_xi, label="i=%s" % (i + 1))
    plt.legend()
```

```python
def plot_convergence_z(A, x_history, x_star, time_avg=False):
    if time_avg:
        x_history = time_average(x_history)

    U, S, Vt = np.linalg.svd(A)
    x_history = x_history.T  # now it's a (N x T) matrix where each column is an x_t
    z_history = Vt @ x_history  # project each x_t onto the V basis
    z_star = Vt @ x_star

    plt.xlabel("$t$")
    plt.ylabel(r"$|v_{i}^{\top}(x_{t} - x_{\lambda}^{\star})|$")
    plt.yscale("log")
    for i in range(A.shape[1]):
        dist_zi = np.abs(z_history[i] - z_star[i])
        plt.plot(dist_zi, label=r"i=%s, $\sigma_{%s} = %s$" % (i + 1, i + 1, round(S[i], 2)))
    plt.legend()
```

```python
def plot_convergence_f(f, x_history, x_star, time_avg=False):
    if time_avg:
        x_history = time_average(x_history)

    f_history = np.array([f(x_history[i]) for i in range(len(x_history))])
    f_star = f(x_star)

    plt.xlabel("$t$")
    plt.ylabel(r"$|f_{\lambda}(x_{t}) - f_{\lambda}(x_{\lambda}^{\star})|$")
    dist_f = np.abs(f_history - f_star)
    plt.yscale("log")
    plt.plot(dist_f)
```

```python
def convergence_heatmap(A, y, etas, lmbdas, gd_fn, T, convergence_tol, time_avg):
    eta_lmbda_matrix = np.zeros(shape=(len(etas), len(lmbdas)))
    for i in range(len(etas)):
        for j in range(len(lmbdas)):
            x_history = gd_fn(A, y, lmbdas[j], etas[i], T)
            if time_avg:
                x_history = time_average(x_history)
            x_T = x_history[-1]
            x_rr = ridge_regression_sol(A, y, lmbdas[j])
            eta_lmbda_matrix[i, j] = 1.0 if np.linalg.norm(x_rr - x_T) < convergence_tol else 0.0

    plt.xlabel(r"$\lambda$")
    plt.ylabel(r"$\eta$")

    plt.xticks(range(len(lmbdas)), lmbdas, rotation=45)
    plt.yticks(range(len(etas)), etas)

    plt.imshow(eta_lmbda_matrix, aspect='auto', origin='lower')
    cbar = plt.colorbar()
    cbar.set_ticks([0, 1])
    cbar.set_ticklabels(["Does not \n converge", "Converges"])

def convergence_heatmap_gd(A, y, etas, lmbdas, T=200, convergence_tol=0.1, time_avg=False):
    convergence_heatmap(A, y, etas, lmbdas, ridge_regression_gd, T, convergence_tol, time_avg)

def convergence_heatmap_sgd(A, y, etas, lmbdas, T=200, convergence_tol=0.1, time_avg=False):
    convergence_heatmap(A, y, etas, lmbdas, ridge_regression_sgd, T, convergence_tol, time_avg)
```

```python
# Experiments for report sections (g), (h), (i)
np.random.seed(227)

# Re-run core definitions in case kernel was restarted

def make_f(A, y, lmbda):
    m, n = A.shape
    return lambda x: 0.5 * ((np.linalg.norm(A @ x - y) ** 2) / m + lmbda * np.linalg.norm(x) ** 2)

def ridge_regression_sol(A, y, lmbda):
    m, n = A.shape
    return np.linalg.inv(A.T @ A + lmbda * m * np.eye(A.shape[1])) @ A.T @ y

def sigma_max(A):
    S = np.linalg.svd(A, full_matrices=True, compute_uv=False)
    return S[0]

def sigma_min(A):
    S = np.linalg.svd(A, full_matrices=True, compute_uv=False)
    if S.shape[0] < A.shape[1]:
        return np.float64(0.0)
    return S[-1]

def gd(x0, grad_fn, eta, T):
    x_history = []
    x = x0.copy()
    for _ in range(T):
        x_history.append(x.copy())
        x -= grad_fn(x) * eta
    return np.array(x_history)

def eta_opt(A, lmbda):
    m, n = A.shape
    return 2 * m / (sigma_max(A) ** 2 + sigma_min(A) ** 2 + 2 * lmbda * m)

def ridge_regression_gd(A, y, lmbda, eta, T):
    m, n = A.shape
    def grad_fn(x):
        return A.T @ (A @ x - y) / m + lmbda * x
    x0 = np.random.standard_normal((n,))
    return gd(x0, grad_fn, eta, T)

def ridge_regression_sgd(A, y, lmbda, eta, T):
    m, n = A.shape
    def grad_fn(x):
        idx = np.random.randint(0, m)
        a_i = A[idx]
        return a_i * (a_i @ x - y[idx]) + lmbda * x
    x0 = np.random.standard_normal((n,))
    return gd(x0, grad_fn, eta, T)

def time_average(x_history):
    time_sum = np.cumsum(x_history, axis=0)
    return time_sum / (1 + np.arange(x_history.shape[0]).reshape(-1, 1))


def make_data(n, m=50, noise_std=0.5):
    if n == 1:
        x_star = np.array([3.0])
    elif n == 2:
        x_star = np.array([3.0, 1.0])
    else:
        x_star = np.random.uniform(low=1, high=3, size=(n,))
    A = np.random.uniform(low=-1, high=1, size=(m, n))
    y = A @ x_star + noise_std * np.random.standard_normal((m,))
    return A, y


def run_gd_sweep(A, y, lmbda, T, factors=(0.3, 0.9, 1.1, 1.5)):
    eta_star = eta_opt(A, lmbda)
    x_rr = ridge_regression_sol(A, y, lmbda)
    f = make_f(A, y, lmbda)
    rows = []
    for c in factors:
        eta = c * eta_star
        x_hist = ridge_regression_gd(A, y, lmbda, eta, T)
        last_err = np.linalg.norm(x_hist[-1] - x_rr)
        f_gap = abs(f(x_hist[-1]) - f(x_rr))
        rows.append((c, eta, last_err, f_gap))
    return rows


def run_sgd_sweep(A, y, lmbda, T, factors=(0.05, 0.1, 0.2, 0.5), repeats=6):
    eta_star = eta_opt(A, lmbda)
    x_rr = ridge_regression_sol(A, y, lmbda)
    f = make_f(A, y, lmbda)
    rows = []
    for c in factors:
        eta = c * eta_star
        last_errs, avg_errs, f_last, f_avg = [], [], [], []
        for _ in range(repeats):
            x_hist = ridge_regression_sgd(A, y, lmbda, eta, T)
            x_avg = time_average(x_hist)[-1]
            last_errs.append(np.linalg.norm(x_hist[-1] - x_rr))
            avg_errs.append(np.linalg.norm(x_avg - x_rr))
            f_last.append(abs(f(x_hist[-1]) - f(x_rr)))
            f_avg.append(abs(f(x_avg) - f(x_rr)))
        rows.append((
            c,
            eta,
            float(np.mean(last_errs)),
            float(np.mean(avg_errs)),
            float(np.mean(f_last)),
            float(np.mean(f_avg)),
        ))
    return rows


configs = [
    (1, [0.0, 0.1, 1.0], 400, 4000),
    (2, [0.0, 0.1, 1.0], 600, 6000),
    (10, [0.01, 0.1, 1.0], 1200, 12000),
]

all_results = {}
for n, lmbdas, T_gd, T_sgd in configs:
    A, y = make_data(n)
    all_results[n] = {}
    for lmbda in lmbdas:
        gd_rows = run_gd_sweep(A, y, lmbda, T=T_gd)
        sgd_rows = run_sgd_sweep(A, y, lmbda, T=T_sgd)
        all_results[n][lmbda] = {"gd": gd_rows, "sgd": sgd_rows}

for n in all_results:
    print(f"\n===== n = {n} =====")
    for lmbda, out in all_results[n].items():
        print(f"\n  lambda = {lmbda}")
        print("  GD: c, eta, ||x_T-x*||, |f(x_T)-f(x*)|")
        for row in out["gd"]:
            print("   ", tuple(np.round(row, 6)))
        print("  SGD: c, eta, E||x_T-x*||, E||xbar_T-x*||, E|f(x_T)-f(x*)|, E|f(xbar_T)-f(x*)|")
        for row in out["sgd"]:
            print("   ", tuple(np.round(row, 6)))
```

```python
# Visualization pass for (g), (h), (i): GD vs SGD and time-averaged SGD

def plot_convergence_x(A, x_history, x_star, time_avg=False):
    if time_avg:
        x_history = time_average(x_history)
    x_history = x_history.T
    plt.xlabel("t")
    plt.ylabel("|x_t - x*|_i")
    plt.yscale("log")
    for i in range(A.shape[1]):
        dist_xi = np.abs(x_history[i] - x_star[i])
        plt.plot(dist_xi)


def plot_convergence_z(A, x_history, x_star, time_avg=False):
    if time_avg:
        x_history = time_average(x_history)
    U, S, Vt = np.linalg.svd(A)
    x_history = x_history.T
    z_history = Vt @ x_history
    z_star = Vt @ x_star
    plt.xlabel("t")
    plt.ylabel("|v_i^T(x_t - x*)|")
    plt.yscale("log")
    for i in range(A.shape[1]):
        dist_zi = np.abs(z_history[i] - z_star[i])
        plt.plot(dist_zi)


def plot_convergence_f(f, x_history, x_star, time_avg=False):
    if time_avg:
        x_history = time_average(x_history)
    f_history = np.array([f(x_history[i]) for i in range(len(x_history))])
    f_star = f(x_star)
    dist_f = np.abs(f_history - f_star)
    plt.xlabel("t")
    plt.ylabel("|f(x_t)-f(x*)|")
    plt.yscale("log")
    plt.plot(dist_f)

np.random.seed(227)
vis_configs = [
    (1, 0.1, 300, 3000, 1.1, 0.1),
    (2, 0.1, 500, 5000, 1.1, 0.1),
    (10, 0.01, 1200, 12000, 1.1, 0.05),
]

for n, lmbda, T_gd, T_sgd, gd_factor, sgd_factor in vis_configs:
    A, y = make_data(n)
    x_rr = ridge_regression_sol(A, y, lmbda)
    f = make_f(A, y, lmbda)
    eta_star = eta_opt(A, lmbda)

    x_hist_gd = ridge_regression_gd(A, y, lmbda, gd_factor * eta_star, T_gd)
    x_hist_sgd = ridge_regression_sgd(A, y, lmbda, sgd_factor * eta_star, T_sgd)

    plt.figure(figsize=(15, 8))

    plt.subplot(2, 3, 1)
    plot_convergence_x(A, x_hist_gd, x_rr, time_avg=False)
    plt.title(f"n={n} GD (last iterate)")

    plt.subplot(2, 3, 2)
    plot_convergence_x(A, x_hist_sgd, x_rr, time_avg=False)
    plt.title(f"n={n} SGD (last iterate)")

    plt.subplot(2, 3, 3)
    plot_convergence_x(A, x_hist_sgd, x_rr, time_avg=True)
    plt.title(f"n={n} SGD (time-average)")

    plt.subplot(2, 3, 4)
    plot_convergence_z(A, x_hist_sgd, x_rr, time_avg=False)
    plt.title("SGD in V basis (last iterate)")

    plt.subplot(2, 3, 5)
    plot_convergence_z(A, x_hist_sgd, x_rr, time_avg=True)
    plt.title("SGD in V basis (time-average)")

    plt.subplot(2, 3, 6)
    plot_convergence_f(f, x_hist_sgd, x_rr, time_avg=True)
    plt.title("SGD objective gap (time-average)")

    plt.tight_layout()
    plt.show()
```
